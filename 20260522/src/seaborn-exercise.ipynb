```python
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False)
```

```python
import seaborn as sns

tips = sns.load_dataset('tips')     #인터넷 서버에 저장된 예제를 tips라고 넣어줘
tips.head(10)
```

```python
sns.barplot(data=tips, x='day', y ='tip', estimator='mean', errorbar=None)
# 막대그래프이고 x축은 날짜, y축은 x축은 팁, estimator='mean'은 평균값 계산기
# estimator='mean'은 요일별로 팁을 전부 더한 뒤 인원수로 나눠서 평균값을 계산해서
# 그 평균 금액만큼 막대기 높이를 세우라는 뜻

# errorbar=None : 복잡한 통계용 검은 선은 없애고 깔끔하게 막대기만 보여달라는 명령
plt.show()
```

```python
sns.barplot(data=tips, x='day', y ='tip', estimator='sum', errorbar=None, hue='sex')
# hue= 'sex'는 남성과 여성 막대기로 2개로 쪼개서 그리게 됨
# 자동적으로 파랑과 주황색으로 나뉘어져있기 때문에 자동적으로 구분해야되니까 설명상자를 자동으로 붙여준것
# estimator = 'sum'은 각 요일별로 손님들이 낸 팁의 총금액(합계)를 뜻함

# 요일 안에서 성별로 한번 더 쪼개짐
# 토요일에 전체 손님이 준 팁 평균으로 쪼개고 그 후에 남성손님은 이만큼, 여성손님은 이만큼 이런식으로 나눔

plt.show()
```

palette='pastel' 👉 부드럽고 연한 파스텔톤 색상으로 변경

palette='coolwarm' 👉 차가운 색(파랑)과 따뜻한 색(빨강)의 그라데이션 조합으로 변경

palette='Set2' 👉 조금 더 차분하고 모던한 허브 느낌의 색상 조합으로 변경

palette='husl' 👉 무지개처럼 다채롭고 화려한 색상 조합으로 변경

```python
sns.barplot(data=tips, x='day', y ='tip', estimator='sum', errorbar=None, hue='sex', palette='Set1')
# palette='Set1'은 그래프의 색상을 바꾸어주는 컬러팩 옵션 같은 것

plt.show()
```

```python
sns.barplot(data=tips, x='day', y ='tip', estimator='sum', errorbar=None, hue='time', palette='Set1')
plt.show()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip')
# scatterplot는 산점도(산포도)를 뜻함

plt.show()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='smoker', size='size')
# hue='smoker'는 흡연자와 비흡연자로 구분
# size='size'는 tips 데이터프레임에 있는 '일행 수(인원 수)'열 이름
# 점의 크기는 이 손님이 몇 명의 일행과 함께 왔는지.
plt.show()
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='smoker', size='size', col='time', style='smoker')
# 식사금액과 팁의 관계를 보여주는데, 점의 색상과 모양은 흡연 여부로 구별하고, 점의 크기는 일행 수로 다르게 해줘
# 이 복잡한 그래프를 점심과 저녁 시간대별로 방을 따로 파서 양옆으로 2개의 그래프로 나누어 그려줘

# col='time'은 컬럼의 약자. time열에는 lunch와 dinner라는 2가지 값이 있음
# 왼쪽은 lunch그래프, 오른쪽은 dinner그래프 나열
# style='smoker'은 데이터 그룹에 따라 점의 모양(스타일)을 다르게 바꿔주는 옵션
#흡연자는 x점, 비흡연자는 o점으로 다르게 찍어줌.
plt.show()
```
