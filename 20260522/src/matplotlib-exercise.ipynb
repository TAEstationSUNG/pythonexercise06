```python
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic')      # 맑은 고딕 폰트 설정
plt.rc('axes', unicode_minus=False)         # 마이너스 기호 깨지지않게 설정
```

```python
age = [7, 8, 9, 10, 11]                         # x축에 들어갈 데이터(나이)
height = [122.1, 127.9, 133.4, 138.8, 142.7]    # y축에 들어갈 데이터(키)

plt.plot(age, height, color='red', marker='o', ls=':')
# plt.plot(x, y,(기본적으로 꺾은선그래프)  선과 점의 색상을 빨간색으로
# marker =0 은 데이터값이 위치한 지점(좌표)에 동그라미 모양을 찍어줌
# ls=':'는 선스타일의 약자. 콜론은 선을 점선으로 그리라는 뜻)

# 실선을 하고싶으면 '-' / 대시선은 '--' 으로
plt.show()
```

```python
blood_type = ['A', 'B', 'O', 'AB']      # x축 : 혈액형
count = [45, 35, 15, 5]                 # y축 : 인원 수

plt.bar(blood_type, count, color=['red', 'blue', 'green', 'purple'])
plt.show()
```

```python
plt.barh(blood_type, count, color=['red', 'blue', 'green', 'purple'])
plt.show()

# 가로(옆으로 누운) 막대그래프
# plt.barh()는 h는 가로를 뜻하는 Horizontal
```

```python
plt.pie(count, colors=['red', 'blue', 'green', 'purple'])
plt.show()
```

```python
plt.pie(count, labels=blood_type, colors=['red', 'blue', 'green', 'purple'], autopct='%.1f%%')
plt.show()

# plt.pie(count) : 원형 그래프를 그리는 핵심 명령어
# autopct = '%.1f%%' : auto percent(자동퍼센트계산)의 약자. 소수점 아래 첫째 자리표현
# %.1f : 소수점 아래 첫째 자리까지 표현하라는 뜻
# %% : 글자 맨 뒤에 퍼센트 기호를 붙여서 출력하라는 뜻
```

```python
# 시본 라이브러리를 다운받았음
# 시본 다이어리는 맷플롯립보다 좀 더 괜찮고 판다스 데이터프레임이랑 잘 맞음

import seaborn as sns

tips = sns.load_dataset('tips')     # 시본 라이브러리에 내장데이터 중 tip를 가져옴
tips.head(3)                    # tips표에서 맨 위에 있는 3줄(행)의 데이터만 보여줘
```

```python
# 산점도(산포도)를 그래프를 그리는 코드

x = tips['total_bill']      # x축은 총 식사 금액
y = tips['tip']             # y축은 준 팁 금액

plt.scatter(x, y)       # plt.scatter(x,y)는 산점도를 그리는 명령어
plt.xlabel('total_bill')    # 가로축에 total_bill이라고 이름표 붙이기
plt.ylabel('tip')           # 세로축에 tip 이라고 이름표 붙이기
plt.show()
```

```python
plt.hist(tips['total_bill'])
plt.show()

# tips['total_bill'] : 팁 데이터 중에서 총 식사 금액컬럼만 뽑아낸 데이터
# plt.hist() : 맷플롯핍에서 히스토그램을 그리는 핵심 명령어
```

```python
plt.hist(tips['total_bill'], bins=20)
# bins =20 이면 식사 금액 데이터를 총 20개의 구간으로 나눠달라는 명령

plt.show()
```

```python
plt.boxplot(tips['total_bill'])
# plt.boxplot()는 박스 플롯 만드는 명령어

plt.show()
```

```python
import pandas as pd

df1 = pd.DataFrame({'월':[1, 2, 3, 4, 5], '몸무게':[80, 78, 75, 73, 70]}) #데이터 1번
df2 = pd.DataFrame({'월':[1, 2, 3, 4, 5], '몸무게':[60, 62, 59, 55, 54]}) #데이터 2번
```

```python
plt.plot(df1['월'], df1['몸무게'], label='철수')    #가로축을 월, 세로축을 몸무게로 해서 꺾은선그래프
plt.plot(df2['월'], df2['몸무게'], label='영희')    #가로축을 월, 세로축을 몸무게로 해서 꺾은선그래프
plt.legend()        #legend()는 지도의 범례나 설명상자, 작은 안내 상자를 띄워줌
plt.show()

# 컴퓨터가 알아서 색상을 다르게 칠해줌
```
