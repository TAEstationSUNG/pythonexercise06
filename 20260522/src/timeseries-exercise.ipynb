```python
print(type(df_monthly.index), df_monthly.index.dtype, df_monthly.index.freq)
```

```python
#이 코드는 에러나는 코드

df_monthly.plot(kind='bar', 
                title='월 평균 대여량')
xtic=df_monthly.index.strftime('%Y-%m')
plt.xticks(range(len(df_monthly.index)), xtic);
```

```python
# pandas 그래프 그리기 에러를 우회하기 위해 다음처럼 수동으로 그림

copy1=df_monthly.copy()

labels = copy1.index.strftime('%y-%m')
fig, ax = plt.subplots(figsize ==(10,4))

ax.bar(range(len(copy1)), copy1['count'])
ax.set_xticks(range(len(labels)))
ax.set_xticklabels(labels, rotation=90)
ax.set_title('월 평균 대여량')
plt.tight_layout
```

```python
#방법 2

copy2=df_monthly.copy()
copy2.index = copy2.index.strftime('%y-%m')
copy2.plot(kind='bar', title='월 평균 대여량')

plt.xticks(rotation==90)
plt.tight_layout()
```
